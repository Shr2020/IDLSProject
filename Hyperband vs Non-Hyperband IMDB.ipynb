```python
import numpy as np
import seaborn as sns

models = ['CNN', 'LSTM', 'QRNN', 'CNN-LSTM', 'LSTM-CNN']
cnn_train_time = cnn_df['train_time'].tolist()
lstm_train_time = lstm_df['train_time'].tolist()
qrnn_train_time = qrnn_df['train_time'].tolist()
cnn_lstm_train_time = cnn_lstm_df['train_time'].tolist()
lstm_cnn_train_time = lstm_cnn_df['train_time'].tolist()

cnn_train_time_hb = cnn_hb_df['train_time'].tolist()
lstm_train_time_hb = lstm_hb_df['train_time'].tolist()
qrnn_train_time_hb = qrnn_hb_df['train_time'].tolist()
cnn_lstm_train_time_hb = cnn_lstm_hb_df['train_time'].tolist()
lstm_cnn_train_time_hb = lstm_cnn_hb_df['train_time'].tolist()

train_times = [np.array(cnn_train_time).sum(), np.array(lstm_train_time).sum(), np.array(qrnn_train_time).sum(), 
               np.array(cnn_lstm_train_time).sum(), np.array(lstm_cnn_train_time).sum()]

train_times_hb = [np.array(cnn_train_time_hb).sum(), np.array(lstm_train_time_hb).sum(), np.array(qrnn_train_time_hb).sum(), 
               np.array(cnn_lstm_train_time_hb).sum(), np.array(lstm_cnn_train_time_hb).sum()]

df_tt = pd.DataFrame(list(zip(models, train_times, train_times_hb)), columns =['Models', 'Train Time', 'Train Time with Hyperband'], index=models)

ax = df_tt.plot.bar(rot=0)

for container in ax.containers:
    ax.bar_label(container, padding=3, rotation=45)
```
