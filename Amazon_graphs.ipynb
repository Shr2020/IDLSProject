```python
import numpy as np
import seaborn as sns

models = ['CNN', 'LSTM', 'QRNN', 'CNN-LSTM', 'LSTM-CNN']
cnn_train_time = cnn_df['train_time'].tolist()
lstm_train_time = lstm_df['train_time'].tolist()
qrnn_train_time = qrnn_df['train_time'].tolist()
cnn_lstm_train_time = cnn_lstm_df['train_time'].tolist()
lstm_cnn_train_time = lstm_cnn_df['train_time'].tolist()

train_times = [np.array(cnn_train_time).sum(), np.array(lstm_train_time).sum(), np.array(qrnn_train_time).sum(), 
               np.array(cnn_lstm_train_time).sum(), np.array(lstm_cnn_train_time).sum()]

df_tt = pd.DataFrame(list(zip(models, train_times)), columns =['Models', 'TT'])

plt.figure(figsize=(6, 6))
plots = sns.barplot(x="Models", y="TT", data=df_tt)

# Iterrating over the bars one-by-one
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),(bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center', va='center', size=10, xytext=(0, 8), textcoords='offset points')

plt.title("Train Time for CNN, LSTM, QRNN, CNN-LSTM, LSTM-CNN")
plt.xlabel("Models", size=15)
plt.ylabel("Train Time", size=15)
plt.show()


# plt.figure(figsize=(12,6))
# plt.plot(x, cnn_train_time, label='CNN Training time')
# plt.plot(x, lstm_train_time, label='LSTM Training time')
# plt.plot(x, qrnn_train_time, label='QRNN Training time')
# plt.xlabel('Epochs')
# plt.ylabel('Train Time')
# plt.title("Train Time for CNN, LSTM, QRNN")
# plt.legend()
# plt.show()
```

```python
models = ['CNN', 'QRNN', 'LSTM', 'CNN-LSTM', 'LSTM-CNN']
cnn_test_acc = cnn_df['Test_Accuracy'].iloc[-1]
lstm_test_acc = lstm_df['Test_Accuracy'].iloc[-1]
qrnn_test_acc = qrnn_df['Test_Accuracy'].iloc[-1]
cnn_lstm_acc = cnn_lstm_df['Test_Accuracy'].iloc[-1]
lstm_cnn_acc = lstm_cnn_df['Test_Accuracy'].iloc[-1]

test_acc = [cnn_test_acc, qrnn_test_acc, lstm_test_acc, cnn_lstm_acc, lstm_cnn_acc]
df_test_acc = pd.DataFrame(list(zip(models, test_acc)), columns =['Models', 'Test Accuracy'])

plt.figure(figsize=(6, 6))
plots = sns.barplot(x="Models", y="Test Accuracy", data=df_test_acc)

# Iterrating over the bars one-by-one
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),(bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center', va='center', size=10, xytext=(0, 8), textcoords='offset points')

plt.title("Test Accuracy for CNN, LSTM, QRNN, CNN-LSTM, LSTM-CNN")
plt.xlabel("Models", size=15)
plt.ylabel("Test Accuracy", size=15)
plt.show()

# plt.figure(figsize=(6,6))
# plt.bar(models, test_acc)
# plt.xlabel('Models')
# plt.ylabel('Test Accuracy')
# plt.title("Test Accuracy for CNN, LSTM, QRNN")
# plt.legend()
# plt.show()
```
